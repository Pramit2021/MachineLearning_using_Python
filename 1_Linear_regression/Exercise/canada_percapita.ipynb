### Exercise:
##### Predict per capita income of canada in the year 2020 with the help of linear regression

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
### Given below is a file that shows per capita income of canada in the years 1970-2015
percapita_income = pd.read_excel('canada_percapita_USD.xlsx')
percapita_income.head(49)
```

```python
plt.xlabel('Year')
plt.ylabel('Per capita income(US$)')
plt.scatter(percapita_income['year'],percapita_income['per capita income'], color='black', marker='*')
plt.show()
```

```python
# Years
year_only = percapita_income.drop('per capita income', axis='columns')
year_only.head(49)
```

```python
#Per capita incpme
income_only = percapita_income.drop('year', axis='columns')
income_only.head(49)
```

#### Making linearregression

```python
reg = LinearRegression()
```

```python
reg.fit(year_only, income_only)
```

#### Prediction in year 2020

```python
### In year 2020
reg.predict([[2020]])
```

##### RESULT: The per capita income of canada in the year 2020 is 41288.69409442 US$
