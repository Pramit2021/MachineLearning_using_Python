```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import matplotlib.pyplot as plt
%matplotlib inline
digits = load_digits()
```

```python
digits.feature_names
```

```python
digits.target_names
```

```python
digits.target
```

```python
df = pd.DataFrame(digits.data, digits.target)
df.head(1800)
```

```python
df['target'] = digits.target
df.head(1800)
```

```python
x = df.drop('target', axis='columns')
y = df['target']
```

```python
x.head(1800)
```

```python
y.head(1800)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
### kernal = rbf
model_rbf = SVC(kernel='rbf')
model_rbf.fit(x_train,y_train)
```

```python
### Accuracy score with kernel = 'rbf'
model_rbf.score(x_test,y_test)
```

```python
### kernal = linear
model_linear = SVC(kernel='linear')
model_linear.fit(x_train,y_train)
```

```python
### Accuracy score with kernel='linear'
model_linear.score(x_test,y_test)
```

```python
### gamma = 1
model_gamma1 = SVC(gamma=1)
model_gamma1.fit(x_train,y_train)
```

```python
### Accuracy acore for gamma=1
model_gamma1.score(x_test,y_test)
```

```python
### gamma = 10
model_gamma10 = SVC(gamma=10)
model_gamma10.fit(x_train,y_train)
```

```python
### Accuracy with gamma = 10
model_gamma10.score(x_test,y_test)
```

```python
model = SVC()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
