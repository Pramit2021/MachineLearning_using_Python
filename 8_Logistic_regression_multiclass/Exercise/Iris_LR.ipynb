```python
from sklearn.linear_model import LogisticRegression
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = load_iris()
dir(iris)
```

```python
iris.feature_names[0:4]
```

```python
iris.target_names[0:3]
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df['target_name'] = iris.target
df.head(60)
```

```python
# 0 = sestona
# 1 = versicolor
# 2 = virginica
iris.target
```

```python
# x = df[['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']]
# y = df['target_name']

x_train,x_test,y_train,y_test = train_test_split(df[['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']],df['target_name'],test_size=0.2)
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_predicted = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
